```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
from functools import reduce
```

```python
collection = MongoClient().billboard.hot100
```

```python
df = pd.DataFrame([i for i in collection.find()])
```

```python
df = pd.DataFrame(collection.find()).drop(columns=['_id', '_max_retries', '_timeout', 'name', 'nextDate', 'title'])
```

```python
alltracks=[]
for row in df.values:
    date = row[0]
    for track in row[1]:
        track['date'] = date
        alltracks.append(track)
```

```python
df = pd.DataFrame(alltracks)
```

```python
df = df.groupby(['artist', 'title']).agg({'date':min, 'peakPos':max, 'weeks':max}, axis='columns').reset_index()
```
